````python
while True:
  input_user_name = input("Enter user name: ")
  input_password = input("Enter password: ")
  if input_user_name in user_data:
    if input_password == user_data[input_user_name]:
      print("login successfull")
      break
    else:
      print("enter correct password")

  else:
    print("user_name is not correct")
````
